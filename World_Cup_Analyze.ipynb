```python
# Argentina matches from 2000 until now
# First, add a column year in the international matches dataframe by spliting the date and only get year
```

```python
international_mat_df['Year']=[x[:4] for x in international_mat_df['Date']]
```

```python
international_mat_df.info()
```

```python
#Convert object data type in Year to int
international_mat_df['Year'] = pd.to_numeric(international_mat_df['Year'])
```

```python
international_mat_df.info()
```

```python
def win_team(home_goal,away_goal):
    if home_goal > away_goal:
        return 1
    else:
        return 0
```

```python
arg_home_inter_mat_stat = international_mat_df[(international_mat_df['Home Team']=='Argentina')&(international_mat_df['Year']>=2018)]
arg_home_inter_mat_stat['Win Bool']=arg_home_inter_mat_stat.apply(lambda x: win_team(x['Home Goals'],x['Away Goals']),axis=1)
arg_home_inter_mat_stat
```

```python
arg_home_inter_mat_stat.to_csv('arg_win_all_match_2018_2022.csv')
```

```python
# Total winning times that Argentina won under home goals from 2018 - 2022
arg_home_inter_mat_stat[arg_home_inter_mat_stat['Home Goals']>arg_home_inter_mat_stat['Away Goals']].shape[0]
```

```python
arg_away_inter_mat_stat = international_mat_df[(international_mat_df['Away Team']=='Argentina')&(international_mat_df['Year']>=2018)]
arg_away_inter_mat_stat
```

```python
#Total winning times when Argentina is away from 2018 -2022
arg_away_inter_mat_stat[arg_away_inter_mat_stat['Away Goals']>arg_away_inter_mat_stat['Home Goals']].shape[0]
```

### Argentina vs France and France vs Argentina

### Argentina is Home Team

```python
arg_home_int = international_mat_df[(international_mat_df['Home Team']=='Argentina')&(international_mat_df['Away Team']=='France')]
arg_home_int = arg_home_int[['Home Team','Home Goals','Away Goals','Away Team','Year']]
```

```python
arg_home_wc = world_cup_matches_df[(world_cup_matches_df['Home Team']=='Argentina')&(world_cup_matches_df['Away Team']=='France')]
arg_home_wc = arg_home_wc[['Home Team','Home Goals','Away Goals','Away Team','Year']]
```

```python
frames = [arg_home_int,arg_home_wc]
arg_home_total = pd.concat(frames)
```

```python
arg_home_total
```

```python
arg_home_total['Win'] = arg_home_total.apply(lambda x: win_team(x['Home Goals'],x['Away Goals']),axis=1)
```
